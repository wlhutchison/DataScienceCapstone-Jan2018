```python
ordered_rucc = ['Metro - Counties in metro areas of 1 million population or more', 'Metro - Counties in metro areas of 250,000 to 1 million population',
                'Metro - Counties in metro areas of fewer than 250,000 population',
                'Nonmetro - Urban population of 20,000 or more, adjacent to a metro area',
                'Nonmetro - Urban population of 20,000 or more, not adjacent to a metro area',
                'Nonmetro - Urban population of 2,500 to 19,999, adjacent to a metro area',
                'Nonmetro - Urban population of 2,500 to 19,999, not adjacent to a metro area',
                'Nonmetro - Completely rural or less than 2,500 urban population, adjacent to a metro area',
                'Nonmetro - Completely rural or less than 2,500 urban population, not adjacent to a metro area']
```

```python
train_x.area__rucc = train_x.area__rucc.astype("category", ordered=True, categories=ordered_rucc ).cat.codes
```

```python
ordered_urban_influence = ['Large-in a metro area with at least 1 million residents or more',
                           'Small-in a metro area with fewer than 1 million residents',
                           'Micropolitan adjacent to a large metro area',
                           'Noncore adjacent to a large metro area',
                           'Micropolitan adjacent to a small metro area',
                           'Noncore adjacent to a small metro with town of at least 2,500 residents',
                           'Noncore adjacent to a small metro and does not contain a town of at least 2,500 residents',
                           'Micropolitan not adjacent to a metro area',
                           'Noncore adjacent to micro area and contains a town of 2,500-19,999 residents',
                           'Noncore adjacent to micro area and does not contain a town of at least 2,500 residents',
                           'Noncore not adjacent to a metro/micro area and contains a town of 2,500  or more residents',
                           'Noncore not adjacent to a metro/micro area and does not contain a town of at least 2,500 residents']
```

```python
train_x.area__urban_influence
```

```python
train_x.area__urban_influence = train_x.area__urban_influence.astype("category", ordered=True, categories=ordered_urban_influence ).cat.codes
```

```python
train_x.area__urban_influence
```

```python
train_x = pd.get_dummies(train_x,columns=['econ__economic_typology'])
```

```python
train_x = pd.get_dummies(train_x,columns=['yr'])
```

```python
train_x.dtypes
```

```python
train_x.health__pop_per_dentist = train_x.health__pop_per_dentist.fillna( train_x.health__pop_per_dentist.mean() )
train_x.health__pct_excessive_drinking = train_x.health__pop_per_dentist.fillna( train_x.health__pct_excessive_drinking.mean() )
train_x.health__homicides_per_100k = train_x.health__pop_per_dentist.fillna( train_x.health__homicides_per_100k.mean() )
train_x.health__motor_vehicle_crash_deaths_per_100k = train_x.health__pop_per_dentist.fillna( train_x.health__motor_vehicle_crash_deaths_per_100k.mean() )
```

```python
train_x = train_x.interpolate(method='polynomial', order=2)
```

```python
train_x = train_x.dropna(axis=0)
```

```python
train_x
```

```python
train_x.to_csv("C:\\Global Projects\\Data Science Specialization\\DataScienceCapstone-Jan2018\\processed_train_values.csv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.7, random_state=7)
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error: %.2f"
      % mse )
print("RMSE: %.2f"
      % math.sqrt(mse) )
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))

# R2 Score
print("R2 Score: %.2f" % regr.score(X_test, y_test))
```
